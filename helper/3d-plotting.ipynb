```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(42)
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)
c = np.random.rand(num_points)  # Data for coloring

# Create the 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(15,12))

# XY plot
axes[0, 0].scatter(x, y, c=c, cmap='viridis', marker='s')
axes[0, 0].set_xlabel('X Label')
axes[0, 0].set_ylabel('Y Label')
axes[0, 0].set_title('XY Plot')

# XZ plot
axes[0, 1].scatter(x, z, c=c, cmap='viridis', marker='s')
axes[0, 1].set_xlabel('X Label')
axes[0, 1].set_ylabel('Z Label')
axes[0, 1].set_title('XZ Plot')

# YZ plot
axes[1, 0].scatter(y, z, c=c, cmap='viridis', marker='s')
axes[1, 0].set_xlabel('Y Label')
axes[1, 0].set_ylabel('Z Label')
axes[1, 0].set_title('YZ Plot')

# 3D scatter plot
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
scatter = ax4.scatter(x, y, z, c=c, cmap='viridis', marker='s')
ax4.set_xlabel('X Label')
ax4.set_ylabel('Y Label')
ax4.set_zlabel('Z Label')
ax4.set_title('3D Scatter Plot')

# Create a colorbar for the 3D scatter plot
cbar_ax = fig.add_axes([1.05, 0.2, 0.02, 0.6])  # [left, bottom, width, height]
cbar = plt.colorbar(scatter, cax=cbar_ax)
cbar.set_label('Colorbar Label')

# Adjust layout to prevent overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(42)
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)
c = np.random.rand(num_points)  # Data for coloring

# XY plot
plt.figure(figsize=(5, 5))
plt.scatter(x, y, c=c, cmap='viridis', marker='o')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('XY Plot')
plt.colorbar(label='Colorbar Label')
plt.show()

# XZ plot
plt.figure(figsize=(5, 5))
plt.scatter(x, z, c=c, cmap='viridis', marker='o')
plt.xlabel('X Label')
plt.ylabel('Z Label')
plt.title('XZ Plot')
plt.colorbar(label='Colorbar Label')
plt.show()

# YZ plot
plt.figure(figsize=(5, 5))
plt.scatter(y, z, c=c, cmap='viridis', marker='o')
plt.xlabel('Y Label')
plt.ylabel('Z Label')
plt.title('YZ Plot')
plt.colorbar(label='Colorbar Label')
plt.show()

# 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=c, cmap='viridis', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot')
cbar = fig.colorbar(scatter)
cbar.set_label('Colorbar Label')
plt.show()
```
